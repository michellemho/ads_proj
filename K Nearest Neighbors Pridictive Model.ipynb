```python
# row = our test data set (a single row from the cleaned dataFrame)
row = 60332
test = knn_drop_nan.loc[row,:'city']
test1 = test.reshape(1,-1)
```

```python
nn.kneighbors(test1,5)[1][0]
```

```python
knn_model = knn_drop_nan.ix[nn.kneighbors(test1,5)[1][0],]
# arstmade | 0=no 1=yes
# frisked | 0=bo 1=yes
```

```python
knn_model
```

```python
# pridicting arrest made and frisked
no_arstmade = 0
yes_arstmade = 0

for value in knn_model['arstmade']:
    if value == 0:
        no_arstmade +=1
    else:
        yes_arstmade +=1

print no_arstmade
print yes_arstmade
if no_arstmade > yes_arstmade:
    print 'arstmade: NO'
else:
    print 'arstmade: YES'
    
if knn_drop_nan.loc[row,'arstmade'] == 0:
    print 'actual arstmad: NO'
else:
    print 'actual arstmad: YES'

    
no_frisked = 0
yes_frisked = 0
for value in knn_model['frisked']:
    if value == 0:
        no_frisked +=1
    else:
        yes_frisked +=1

print no_frisked
print yes_frisked
if no_frisked > yes_frisked:
    print 'frisked: NO'
else:
    print 'frisked: YES'

if knn_drop_nan.loc[row,'frisked'] == 0:
    print 'actual friskedd: NO'
else:
    print 'actual frisked: YES'
```
